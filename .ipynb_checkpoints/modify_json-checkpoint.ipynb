```python
import os
import subprocess

def apply_jq_to_json_files(directory):
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.json'):
                file_path = os.path.join(root, file)
                temporary_file_path = file_path + '.tmp'
                
                # 构建 jq 命令
                jq_command = f'jq \'del(.distribution)\' "{file_path}" > "{temporary_file_path}"'
                
                try:
                    # 执行 jq 命令
                    subprocess.run(jq_command, shell=True, check=True)
                    
                    # 替换原文件
                    os.replace(temporary_file_path, file_path)
                    print(f'Updated file: {file_path}')
                except subprocess.CalledProcessError as e:
                    print(f'Failed to process file {file_path}: {e}')
                except Exception as e:
                    print(f'An error occurred with {file_path}: {e}')
```

```python
# 设置你的目录路径
directory_path = 'gap_result/tsp'  # 替换成你的文件夹路径
apply_jq_to_json_files(directory_path)
```

```python
import json
import os
import subprocess

failed_path = []

def check_and_remove_distribution(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            data = json.load(file)
        # 检查 'distribution' 键是否存在
        if "distribution" in data:
            # 如果存在，则使用 jq 命令删除
            use_jq_to_remove_key(file_path)
    except json.JSONDecodeError as e:
        print(f"JSON error in {file_path}, assumed 'distribution' key exists: {e}")
        use_jq_to_remove_key(file_path)
    except Exception as e:
        print(f"Error processing {file_path}: {e}")

def use_jq_to_remove_key(file_path):
    temporary_file_path = file_path + '.tmp'
    jq_command = f'jq \'del(.distribution)\' "{file_path}" > "{temporary_file_path}"'
    try:
        subprocess.run(jq_command, shell=True, check=True)
        # 替换原文件
        os.replace(temporary_file_path, file_path)
        print(f'Updated file: {file_path}')
    except subprocess.CalledProcessError as e:
        failed_path.append(file_path)
        print(f'Failed to process file with jq {file_path}: {e}')
    except Exception as e:
        failed_path.append(file_path)
        print(f'Other error with jq on {file_path}: {e}')

def walk_directory_and_modify_json(directory):
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.json'):
                file_path = os.path.join(root, file)
                check_and_remove_distribution(file_path)
                
```

```python
# 替换这个路径为你的目标文件夹路径
directory_path = 'gap_result/tsp'  # 替换成你的文件夹路径
walk_directory_and_modify_json(directory_path)
```
