```python
df_mean = df.groupby(['log(1/r)', 'N'])['gapLB'].median().reset_index()
```

```python
def fitted_gapLB(N,p,slope,intercept,c1,c2):
    return max(-np.sqrt(np.log((2*p+1)**2)/(slope*N+intercept)) + c1*N+c2 - 1,0)

def fit_model(data):
    def model(N_log1r,slope,intercept,c1,c2):
        N, log1r = N_log1r.T
        return -np.sqrt(log1r/(slope*N+intercept)) + c1*N+c2 - 1
    # Extract data for N, log1r, approx
    N_data = data['N'].values
    log1r_data = data['log(1/r)'].values
    gap_data = data['gapLB'].values
    N_log1r_data = np.column_stack((N_data, log1r_data))
    
    # Initial guesses for the parameters
    initial_guesses = [0.5, -0.7, 0.5, 0]
    # Curve fitting
    params, covariance = curve_fit(model, N_log1r_data, gap_data, p0=initial_guesses)

    # Calculate predictions and mean squared error
    predicted_approx = model(N_log1r_data, *params)
    mse = mean_squared_error(gap_data, predicted_approx)
    
    return params,mse
```

```python
params,mse = fit_model(df_mean[df_mean['gapLB']>0])
print(params,mse)
```
